# matplotlib库

## 引用

`from matplotlib import pyplot as plt`

## 绘图类型

`plt.xxx()`

| 函数名称  | 描述                                         |
|:----------|:---------------------------------------------|
| Bar       | 绘制条形图                                   |
| Barh      | 绘制水平条形图                               |
| Boxplot   | 绘制箱型图                                   |
| Hist      | 绘制直方图                                   |
| his2d     | 绘制2D直方图                                 |
| Pie       | 绘制饼状图                                   |
| Plot      | 在坐标轴上画线或者标记                       |
| Polar     | 绘制极坐标图                                 |
| Scatter   | 绘制x与y的散点图                             |
| Stackplot | 绘制堆叠图                                   |
| Stem      | 用来绘制二维离散数据绘制（又称为“火柴图”） |
| Step      | 绘制阶梯图                                   |
| Quiver    | 绘制一个二维按箭头                           |

## 笔记

### 示例数据

```python
x = [i for i in range(1,21)]
y = [i for i in range(10,210,10)]
print(x)
print(y)
```

### plt.plot(x,y)参数传入x,y 即可生成最原始图表

```python
from matplotlib import pyplot as plt
plt.plot(x,y)
plt.show()
```

### 坐标名称和图名

+ %matplotlib inline 是 Jupyter 提供的魔法命令，它可以把输出图显示在笔记本内部，否则会以查看器的形式单独显示。

#### 函数xlabel、ylabel、title

```python
from matplotlib import pyplot as plt
plt.plot(x,y)
plt.xlabel("x")  # x标题
plt.ylabel("y")  # y标题
plt.title("title")  # 图表标题
plt.show()
```

#### figure对象

+ figure ”图形“，调用对象内部方法来更改画布

+ plt.figure(参数)

| 参数      | 说明                                                   |
| :---------| :------------------------------------------------------|
| figsize   | 指定画布的大小，(宽度,高度)，单位为英寸。              |
| dpi       | 指定绘图对象的分辨率，即每英寸多少个像素，默认值为80。 |
| facecolor | 背景颜色。                                             |
| dgecolor  | 边框颜色。                                             |
| frameon   | 是否显示边框。                                         |


```python
from matplotlib import pyplot as plt

plt.plot(x,y)

plt.xlabel("x")  # x标题
plt.ylabel("y")  # y标题
plt.title("title")  # 图表标题

plt.figure(figsize=(50,40),dpi= 80,facecolor = "red",edgecolor="b",frameon=False)


plt.show()
```
