```python
import json
import requests
import time
import xlwings as xw
import pandas as pd
from pandas import DataFrame

# http://www.cncyms.cn/index.php?m=content&c=index&a=lists&catid=57#
```

```python
reltime = "蔬菜"

url = "http://www.cncyms.cn/pages.php"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36"}
data = {"pageNum":0,
   "pname":"",
   "reltime":reltime,
   }
res = requests.post(url=url,headers = headers,data = data)
js = res.json()
totalPage = js["totalPage"]
df = DataFrame(data = js["list"])
df["ReleaseTime"] = pd.to_datetime(df["ReleaseTime"])

for i in range(2,totalPage+1):
    pageNum = i
    data = {"pageNum":i-1,
   "pname":"",
   "reltime":reltime,
   }
    res = requests.post(url=url,headers = headers,data = data)
    js = res.json()
    df_1 = DataFrame(data = js["list"])
    df_1["ReleaseTime"] = pd.to_datetime(df_1["ReleaseTime"])
    df = pd.concat([df,df_1])
    time.sleep(1)
    if i == 5:
        print(df.head(20))
        break
```

```python
def spide_one_page(pageNum,reltime):
    """
    将每一行数据作为列表返回
    """
    url = "http://www.cncyms.cn/pages.php"
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36"}
    data = {"pageNum":pageNum-1,
       "pname":"",
       "reltime":reltime,
       }
    res = requests.post(url=url,headers = headers,data = data)
    all_js = res.json()
    total = all_js["total"]
    pageSize = all_js["pageSize"]
    totalPage = all_js["totalPage"]
    if pageNum<=totalPage+1:
        for dic in all_js["list"]:
            ls = list(dic.values)
            yield ls
```

```python
def spide_all(star_time,reltime):
    totalPage = int(spide_one_page(1,reltime)[0])+1
    all_info = []
    for pageNum in range(1,totalPage+1):
        one_page_info=spide_one_page(pageNum,reltime)[1]
        all_info.append(one_page_info)
        if one_page_info[-1]["ReleaseTime"] == star_time:
            break
        time.sleep(1)
    return all_info
    
```
